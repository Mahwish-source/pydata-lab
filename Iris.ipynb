```python
# Task 1: Exploring and visulizing a simple data set
# Data set : Iris Data set

# step 1: Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# step 2: Load the data set
# Load directly from Seaborn (built-in)
iris = sns.load_dataset("iris")
```

```python
# Step 3: Inspect the dataset
print("Shape of datset:", iris.shape)
print("\nColumn names:", iris.columns)
print("\nFirst 5 rows:")
print(iris.head())
```

```python
# Step 4: Summary statistics
print("\nSummary statistics:")
print(iris.describe())
```

```python
# Step 5: Data Visualization
# (A) Scatter Plot: Relationship between Sepal Length and Sepal Width
plt.figure(figsize=(6,4))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
# (B) Histogram: Distribution of Sepal Length
plt.figure(figsize=(6,4))
sns.histplot(iris['sepal_length'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# (C) Box Plot: Sepal Length by Species
plt.figure(figsize=(6,4))
sns.boxplot(x='species', y='sepal_length', hue='species', data=iris, palette='Set2', legend=False)
plt.title('Sepal Length by Species')
plt.show()
```
