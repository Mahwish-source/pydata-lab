```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
inputs = df.drop('Survived',axis='columns')
target = df.Survived
```

```python
inputs.Sex = inputs.Sex.map({'male': 1, 'female': 2})
```

```python
dummies = pd.get_dummies(inputs.Sex)
dummies.head(3)
```

```python
inputs = pd.concat([inputs,dummies],axis='columns')
inputs.head(3)
```

```python
# inputs.drop(['Sex','male'],axis='columns',inplace=True) # Original line
# inputs.head(3)

# Drop the original 'Sex' column
if 'Sex' in inputs.columns:
    inputs.drop('Sex', axis='columns', inplace=True)

# Now check for and drop the 'male' column if it exists
if 'male' in inputs.columns:
    inputs.drop('male', axis='columns', inplace=True)
elif 'female' in inputs.columns:
    # If 'male' is not there, maybe the user intended to drop 'female' instead
    # To represent gender with one column, drop one of the dummy columns.
    # If 'male' is not present, it implies 'female' might be, so let's drop 'female'
    # This ensures we represent gender with a single dummy column
    inputs.drop('female', axis='columns', inplace=True)


inputs.head(3)
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age[:10]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head(6)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.3)
```

```python
# Convert all column names to strings
inputs.columns = inputs.columns.astype(str)

# The rest of your code remains the same
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.3)

from sklearn.naive_bayes import GaussianNB
model = GaussianNB()

# Now fit the model with the corrected column names
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
